## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_type = purchase_data_df.loc[:,["SN", "Age"]] 

bins = [0,10, 14, 19, 24, 29, 34, 39, 100]
age_groups = ['< 10', '10-14','15-19','20-24','25-29','30-34','35-39','40+']
real_total = age_type[age_type.duplicated(keep = 'first') == False]

real_total[""]= pd.cut(age_type["Age"], bins, labels = age_groups)
new_real = real_total.groupby('').count()
age_players = pd.DataFrame(new_real)
age_players['Percentage of Players'] = (age_players['Age']/ age_players['Age'].sum())*100

df1 =  age_players.drop(['SN'],axis = 1)
df2 = df1.rename(columns = {'Age':"Total Count"})
df2.round(2)
```

```python
age_price =purchase_data_df.loc[:, ['Age','Price',"Purchase ID"]]
age_price["Range"]= pd.cut(age_type["Age"], bins, labels = age_groups)

test = age_price.loc[:,["Range", "Price"]]
why = test.groupby('Range').sum()

test_2 = age_price.loc[:,["Range","Age"]]
nott = test_2.groupby('Range').count()

why_not_lol = pd.merge(why, nott, on="Range")
purchase_sum = pd.DataFrame(why_not_lol)
purchase_summ = purchase_sum.rename(columns = {'Age':"Purchase Count"})
purchase_summery = purchase_summ.rename(columns = {'Price':"Total Purchase Value"})
purchase_summery["Average Purchase Price"] = (purchase_summery['Total Purchase Value']/ purchase_summery['Purchase Count'])
correct = purchase_summery
correct['Avg Total Purchase per Person'] = (purchase_summery['Total Purchase Value']/df2["Total Count"])
correct.round(2)


```
